```python
from bs4 import BeautifulSoup
import requests
```

```python
#get the source code as a html file
source  = requests.get('https://www.panet.co.il/article/3418634').text 

soup = BeautifulSoup(source,'html')

article = soup.find('article') 
#test with the header of the comment section 
 
comments = article.find('aside',class_='panet-comments')

print(comments.prettify())
```

```python
#first headline of the comment
headline = comments.find('a',class_='panet-title').text

print(headline)
#the comment 

comment = comments.find('div',class_="panet-comment").p

print(comment)
```

```python
#method to show hidden comments (style : Display =block/none)

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup

import time 

#1st try 
class Demo():
    def dem_enable_display(self):

        options  = webdriver.ChromeOptions()

        driver = webdriver.Chrome(executable_path=ChromeDriverManager().install(),options=options)
        
        driver.set_window_size(1120,1000)

        driver.get("https://www.panet.co.il/article/3418634")
    #find the content of the hidden comment by xpath and check if it's displayed
        elem = driver.find_element_by_xpath('//*[@id="articleGrid"]/article/aside[2]/div[2]/div[1]/div[2]').is_displayed()

        print(elem)
        
    #find the "button" and click on it to show the comment
        driver.find_element_by_xpath('//*[@id="articleGrid"]/article/aside[2]/div[2]/div[1]/a').click()

        elem2 = driver.find_element_by_xpath('//*[@id="articleGrid"]/article/aside[2]/div[2]/div[1]/div[2]').is_displayed()

        print(elem2)

        time.sleep(3)

        #xpath of the "2nd button" : //*[@id="articleGrid"]/article/aside[2]/div[2]/div[2]/a
        #xpath of the content : //*[@id="articleGrid"]/article/aside[2]/div[2]/div[2]/div[2]
        
        
    #use beautifulsoup to extract the relevant content
    
        source = driver.page_source
        soup = BeautifulSoup(source,'html.parser')
        article = soup.find('article') 

        #test with the header of the comment section 
        comments = article.find('aside',class_='panet-comments')
        #first headline of the comment
        headline = comments.find('a',class_='panet-title').text

        print(headline)
        #the comment 

        comment = comments.find('div',class_="panet-comment").p.text

        print(comment)

DemoDisplay=Demo()
DemoDisplay.dem_enable_display()
```
